### Image Processing With scikit-image

```python
from skimage import io
from matplotlib import pyplot as plt
```

```python
img = io.imread("../images/image-5.jpg")
```

```python
from skimage.transform import rescale,resize,downscale_local_mean
```

```python
rescaled_img = rescale(img,1/4,anti_aliasing=True)
```

```python
plt.imshow(rescaled_img)
```

#### Edge Detection

```python
from skimage.filters import roberts, sobel,scharr, prewitt
```

```python
img_2 = io.imread("../images/image-7.jpg",as_gray=True)
```

```python
edge_roberts = prewitt(img_2)
plt.imshow(edge_roberts,cmap='gray')
```

```python
from skimage.feature import canny
```

```python
edge_canny = canny(img_2,sigma=4)
plt.imshow(edge_canny)
```

```python
from skimage import restoration
import numpy as np
```

```python
psf = np.ones((3,3)) /9

deconvolved,_ = restoration.unsupervised_wiener(img_2,psf)
plt.imshow(deconvolved)
```
