#### Working with pillow for images

#### Working with matplotlib for images

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
img = mpimg.imread("../images/image-2.jpg")

#By default img is imported as numpy array
print(type(img))
print(img.shape)
```

```python
#showing the image
plt.imshow(img)
plt.colorbar()
```

#### Using sckit-image

```python
from skimage import io, img_as_float, img_as_ubyte
```

```python
image = io.imread("../images/image-2.jpg")
print(type(image))
```

```python
plt.imshow(image)
```

```python
image_flt = img_as_float(image)
print(image_flt)

#Note the values are between 0 and 1 i.e. normalized by dividing by 255
```

```python
img_as_flt_np = io.imread("../images/image-2.jpg").astype(np.float64)
print(img_as_flt_np)

#Note the values are not between 0 and 1 just the integer values
```

#### Using openCV

```python
import cv2
```

```python
img_ocv = cv2.imread("../images/image-3.jpg",1) 
#0 means grey image
#1 means colored image
cv2.imshow("Grey Image",img_ocv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

By default, cv2 handles images as BGR (not RGB)

```python
#Converting to RGB
img_bgr = cv2.imread("../images/image-4.jpg")

img_rgb = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)
cv2.imshow("RGB Image",img_rgb)
cv2.imshow("BGR Image",img_bgr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
